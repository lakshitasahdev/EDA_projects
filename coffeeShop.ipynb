```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
formula = 'Calories ~ C(Milk)'
model = ols(formula,menu).fit()
table = anova_lm(model)
table
```

```python
formula = 'TotalFat ~ C(Milk)'
model = ols(formula,menu).fit()
table = anova_lm(model)
table
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import  scipy.stats as stats
```

```python
formula = 'Calories ~ Sugars + TotalFat + Protein'
reg = smf.ols(formula = formula, data = menu).fit()
print(reg.summary())
```

```python
pred = reg.predict(menu)
```

```python
sns.regplot(pred,menu.Calories)
```
